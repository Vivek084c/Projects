```python
import numpy as np
```

```python

class Layers_Dense:
    def __init__(self, NoOfInputs, NoOfNeurons):
        """
        NoOfInputs : no of inputs to this neuron
        NoOfNeurons : no of neurons in the layer
        """
        self.weights = np.random.randn(NoOfInputs, NoOfNeurons)  * 0.01 # to have small weights initially
        self.g_weights = np.random.randn(NoOfInputs, NoOfNeurons) * 0.01 # to have small weights initially
        """defined as noofinputs x noofneurons 
        to prevent the use of tranoose"""
        self.bias = np.zeros((1, NoOfNeurons))

    def forward(self, inputs):
        self.inputs = inputs    #store the inputs for backpropagation
        self.output = np.dot(inputs, self.weights) + np.dot(inputs, self.g_weights)+ self.bias
        return self.output
    
    def backward(self, dvalues):
        self.dweights = np.dot(self.inputs.T, dvalues)
        self.dbias = np.sum(dvalues, axis=0, keepdims=True)
        self.dinputs = np.dot(dvalues, self.weights.T)
        return self.dinputs
```

```python



class neuron:
    def __init__(self, no_of_inputs):
        self.w = np.random.randn(no_of_inputs)
        self.g = np.random.randn(no_of_inputs)
        self.bias = 0

    def forward(self, inputs):
        self.inputs = inputs
        self.output = np.dot(self.w, inputs) + np.dot(self.g, inputs) + self.bias
        return self.output

    def backward(self, dL_dout, learning_rate): 
        for i in range(len(self.inputs)):
            self.w[i] -= self.inputs[i] * dL_dout * learning_rate
            self.g[i] -= self.inputs[i] * dL_dout * learning_rate
        self.bias -= dL_dout * learning_rate
        return self.w
        
```
